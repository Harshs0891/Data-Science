***Q1.*** Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

5 plots :-
1. Scatter Plot :- used to plot between two variables numerical vs numerical
2. Histogram Plot using displot :- used to visualize single variable
3. Bar Plots :- used to plot categorical data vs numerical data
4. Box Plots :- display the distribution of data including median,first quartile,third quartile
5. HeatMaps :- Ideal for visualizing the relationships between two categorical variables by color-coding the data cells based on the frequency or values of the intersections. They are also used for correlation matrices and showing data density in 2D.

***Q2.*** Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
sns.scatterplot(x=fmri.timepoint,y=fmri.signal)
```

***Q3.*** Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x=titanic.pclass,y=titanic.age)
```

```python
sns.boxplot(x=titanic.pclass,y=titanic.fare)
```

***Q4.*** Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
sns.displot(diamonds['price'])
```

```python
sns.relplot(x=diamonds.x,y=diamonds.y,data=diamonds,hue='cut')
```

***Q5.*** Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris)
```

```python
sns.relplot(x=iris.sepal_length,y=iris.sepal_width,data=iris,hue='species')
```

***Q6.*** Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights
```

```python
x = flights.pivot_table(index='year',columns='month',values='passengers')
```

```python
sns.heatmap(x)
```
