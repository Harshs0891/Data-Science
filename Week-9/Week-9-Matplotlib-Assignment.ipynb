***Q2.*** What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Scatter plot is a type of visualizations used to plot the relation between 2 numerical data individually on a graph through collection of points.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.title("This is the Title of the plot")
plt.xlabel("This is X-axis Data")
plt.ylabel("This is Y-axis Data")
plt.show()
```

The `subplot` function in Matplotlib is used to create a grid of subplots within a single figure, allowing you to display multiple plots or visualizations together in a structured layout. This function helps organize and arrange different plots within a single figure, making it easier to compare and analyze data from multiple sources or angles.

```python
import numpy as np
# For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y,c='r')

# For line 2: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y,c='b')

# For line 3: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y,c='g')

# For line 4: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y,c='y')
```

***Q4.*** What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a data visualization that represents categorical data with rectangular bars, where the length or height of each bar corresponds to the value of a specific category or variable.It is used to create visualization for categorical data and its related numerical values.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

***Q5.*** What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays a summary of key statistics, including the median, quartiles, and potential outliers, using a rectangular "box" and "whiskers" extending from it. Box plots are useful for visualizing the spread and skewness of data and identifying potential outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.show()
```
